```python

fig, ax = plt.subplots(1, 1, figsize=(16, 10), facecolor='white')

df = pd.read_csv('local_data/SF_misID.csv')
df['flav_diff'] = np.abs(df.SF_q2MisID - df.SF_b2MisID)
print(df.flav_diff)

x = df.bin_pT1
y_nom, y_up, y_down = df.SF_b2MisID, df.SF_b2MisID + df.sigma_SF_b2MisID,  df.SF_b2MisID - df.sigma_SF_b2MisID
ax.plot(x, y_nom, 'C0', drawstyle='steps-post', label='_nolegend_')
ax.fill_between(x, y_up, y_down, color = 'C0', alpha=0.9, step='post', hatch='-', label=r'$\sf b\rightarrow\tau$')
#ax.plot(x, y_nom, 'C0', label=r'$\sf b\rightarrow\tau$')
#ax.fill_between(x, y_up, y_down, color = 'C0', alpha=0.5, label='_nolegend_')

y_nom, y_up, y_down = df.SF_q2MisID, df.SF_q2MisID + df.sigma_SF_q2MisID,  df.SF_q2MisID - df.sigma_SF_q2MisID
ax.plot(x, y_nom, 'C1', drawstyle='steps-post', label='_nolegend_')
ax.fill_between(x, y_up, y_down, color = 'C1', alpha=0.9, step='post', hatch='|', label=r'$\sf udscg\rightarrow\tau$')

y_nom, y_up, y_down = df.SF_j2MisID, df.SF_j2MisID + df.sigma_SF_j2MisID,  df.SF_j2MisID - df.sigma_SF_j2MisID
ax.plot(x, y_nom, 'C4', drawstyle='steps-post', label='_nolegend_')
ax.fill_between(x, y_up, y_down, color = 'C4', alpha=0.9, step='post', label=r'$\sf jet\rightarrow\tau$')

ax.plot([20, 65], [1, 1], 'r--', label='_nolegend_')

ax.legend(fontsize=18)
#ax.set_title('$\sf jet\rightarrow \tau$ mis-ID rate')
ax.set_yscale('linear')
ax.set_xlabel(r'$p_{T,j}$')
ax.set_xlim(20., 65)
ax.set_ylabel(r'$\sf jet\rightarrow\tau$ rate')
ax.set_ylim(0.5, 1.5)
ax.grid()

#if ix%10 == 0:
#    ax.set_ylabel('Entries')
#
#ax = axes[2*(ix//5)+1, ix%5]
#x = bins
#y_up = h_plus/h_nominal
#y_down = h_minus/h_nominal
#y_up = np.append(y_up, y_up[-1])
#y_down = np.append(y_down, y_down[-1])
#ax.plot(x, y_up, 'C0', drawstyle='steps-post')
#ax.plot(x, y_down, 'C2', drawstyle='steps-post')
#ax.fill_between(x, y_up, y_down, color = 'C1', alpha=0.5, step='post')
#ax.plot(brange, [1, 1], 'C1--')
#ax.set_ylim(0.8, 1.2)
#ax.grid()

plt.show()
```
