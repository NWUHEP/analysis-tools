```python
# imports and configuration
%cd '/home/naodell/work/CMS/z_plus_jpsi'
import numpy as np
import pandas as pd
```

```python
input_dir  = f'data/templates/bjet_binned_test/'
selections = ['mumu', 'mutau', 'mu4j', 'emu', 'ee', 'etau', 'e4j']
#selections = ['mumu', 'mutau', 'emu']
targets    = dict(
                  mumu  = 'lepton2_pt', 
                  ee    = 'lepton2_pt', 
                  emu   = 'lepton2_pt', #'dilepton1_pt_asym', 
                  mutau = 'lepton2_pt', 
                  etau  = 'lepton2_pt', 
                  mu4j  = 'lepton1_pt',
                  e4j   = 'lepton1_pt'
                 )
decay_map = pd.read_csv('data/decay_map.csv').set_index('id')['decay'].values
names = []
data = []
for sel in selections:
    target = targets[sel]
    for b in [0, 1]:
        df_templates = pd.read_csv(f'{input_dir}/{sel}/{target}_bin-{b}_val.csv').set_index('bins')
        df = df_templates[decay_map]
        s = df.sum(axis=0)
        names.append(f'{sel}_{b}')
        data.append(s/s.sum())

df = pd.concat(data, axis=1)
df.columns = names
#df = df.transpose()
df.to_csv('data/channel_breakdowns.csv')
pd.options.display.float_format = '{:.1f}%'.format
df*100
```
