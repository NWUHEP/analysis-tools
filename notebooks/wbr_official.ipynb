```python
import numpy
import matplotlib.pyplot as plt
import matplotlib

%cd '/home/naodell/work/CMS/z_plus_jpsi'
np.set_printoptions(precision=3)
matplotlib.style.use('default')
params = {'legend.fontsize': 20,
          'axes.labelsize': 20,
          'figure.figsize': (8, 8),
          'axes.facecolor': 'white',
          'axes.titlesize':'x-large',
          'xtick.labelsize':18,
          'ytick.labelsize':18,
         }
matplotlib.rcParams.update(params)
%matplotlib inline

fig = plt.figure(figsize=(8, 8), facecolor='white')
ax = plt.subplot()
br = [10.71, 10.63, 11.38]
br_err = [0.16, 0.15, 0.21]

ax.errorbar(br, [3, 2, 1], xerr = br_err, 
            fmt = 'ko',
            ecolor = 'k',
            capsize = 0.,
            elinewidth = 2.,
            alpha = 0.9
           )

ax.plot([10.86, 10.86], [0, 4], c='C3', linestyle='--')
ax.fill_between([10.86-0.09, 10.86+0.09], [0, 0], [4, 4], color = 'b', alpha = 0.25 )
ax.fill_between([10.86-2*0.09, 10.86+2*0.09], [0, 0], [4, 4], color = 'b', alpha = 0.25 )

p_labels = [r'$\sf W\rightarrow \tau$', r'$\sf W\rightarrow \mu$', r'$\sf W\rightarrow e$'] 
ax.set_yticks([1, 2, 3])
ax.set_yticklabels(p_labels)

ax.set_xlim(10.2, 12.25)
ax.set_ylim(0.5, 3.5)
ax.set_xlabel(r'$\sf \Gamma_{W\rightarrow \ell}/\Gamma_{W}$')
ax.set_title('W leptonic decay modes')

ax.text(11.65, 2.975, r'$(10.71 \pm 0.16)\%$', fontsize=16)
ax.text(11.65, 1.975, r'$(10.63 \pm 0.15)\%$', fontsize=16)
ax.text(11.65, 0.975, r'$(11.38 \pm 0.21)\%$', fontsize=16)

plt.tight_layout()
plt.grid(linestyle='--', axis='x')
plt.savefig(f'plots/w_br_pdg.pdf')
plt.show()
```
