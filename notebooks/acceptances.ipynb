```python
%cd '/home/naodell/work/z_plus_jpsi'

import pandas as pd
import ROOT as r

import scripts.plot_tools as pt
```

```python
selection = 'emu'
datasets = ['ttbar_inclusive', 't_tw', 'tbar_tw']
features = ['n_bjets', 'n_taus', 'gen_cat']
dm = pt.DataManager(input_dir     = f'data/flatuples/{selection}_2016',
                    dataset_names = datasets,
                    selection     = selection,
                    period        = 2016,
                    scale         = 36.4e3,
                    cuts          = 'n_bjets >= 1'
                   )
dataframes = dm.get_dataframes(['ttbar', 't'])
df_ttbar = dataframes['ttbar'][features]
df_t     = dataframes['t'][features]
```

```python
# files for generated event counts
root_file = r.TFile(f'data/bltuples/output_{selection}_2016.root')
hcounts = {d:r.TH1D(root_file.Get(f'GenCategory_{d}')) for d in datasets} # there's a fucking typo!!!
df_counts = pd.DataFrame({d:[h.GetBinContent(i) for i in range(h.GetNbinsX()+1)] for d, h in hcounts.items()})
df_counts['t'] = df_counts['t_tw'] + df_counts['tbar_tw']
df_counts = df_counts.drop(['t_tw', 'tbar_tw'], axis=1)
#df_counts.iloc[:21]
```

```python
# ttbar numbers
cut_list = [df_ttbar.query('n_bjets == 1').gen_cat.value_counts(), df_ttbar.query('n_bjets >= 2').gen_cat.value_counts()]
df = pd.concat([df_counts.ttbar_inclusive] + cut_list, axis=1)
df.columns = ['generated', 'n_bjets = 1', 'n_bjets >= 2']
df = df[1:22]
df.transpose().to_csv(f'data/ttbar_acc_{selection}.csv')
df
```

```python
# t numbers
cut_list = [df_t.query('n_bjets == 1').gen_cat.value_counts(), df_t.query('n_bjets >= 2').gen_cat.value_counts()]
df = pd.concat([df_counts.t] + cut_list, axis=1)
df.columns = ['generated', 'n_bjets = 1', 'n_bjets >= 2']
df = df[1:22]
df.transpose().to_csv(f'data/t_acc_{selection}.csv')
df
```
